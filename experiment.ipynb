# Testing Solution A

```python
n = 3
best_time_list = []
total_time_list = []
train_accs_list = []
val_accs_list = []
test_accs_list = []
for i in range(n):
    _, best_time, total_time, train_accs, val_accs, test_accs = SolA('H.3')
    best_time_list.append(best_time)
    total_time_list.append(total_time)
    train_accs_list.append(train_accs)
    val_accs_list.append(val_accs)
    test_accs_list.append(test_accs)

best_time_avg = np.mean(best_time_list)
best_time_std = np.std(best_time_list)
total_time_avg = np.mean(total_time_list)
total_time_std = np.std(total_time_list)

best_train = []
best_val = []
best_test = []
for i in range(n):
    val_accs = val_accs_list[i]
    index = val_accs.index(max(val_accs))
    best_train.append(train_accs_list[i][index])
    best_val.append(val_accs_list[i][index])
    best_test.append(test_accs_list[i][index])

best_train_avg = np.mean(best_train)
best_train_std = np.std(best_train)
best_val_avg = np.mean(best_val)
best_val_std = np.std(best_val)
best_test_avg = np.mean(best_test)
best_test_std = np.std(best_test)

print('Total Time: %.4f +/- %.4f, Best Model Time: %.4f +/- %.4f, Train Acc: %.4f +/- %.4f, Val Acc: %.4f +/- %.4f, Test Acc: %.4f +/- %.4f' %(
    total_time_avg,
    total_time_std,
    best_time_avg,
    best_test_std,
    best_train_avg,
    best_train_std,
    best_val_avg,
    best_val_std,
    best_test_avg,
    best_test_std
))
```

# Comparing Solutions

```python
_,_,_,time_list_B,train_list,val_list_B,test_list = SolB('H.3')
_,_,_,time_list_C,train_list,val_list_C,test_list = SolC('H.3')
model,_,_,_,_,_,_ = SolC('H.2')
_,_,_,time_list_D_H2,train_list,val_list_D_H2,test_list = FineTune(copy.deepcopy(model),'H.3')
model,_,_,_,_,_,_ = SolC('C.2')
_,_,_,time_list_D_C2,train_list,val_list_D_C2,test_list = FineTune(copy.deepcopy(model),'H.3')
model,_,_,_,_,_,_ = SolC('F.2')
_,_,_,time_list_D_F2,train_list,val_list_D_F2,test_list = FineTune(copy.deepcopy(model),'H.3')

# graph by epoch
epochs = list(range(100))
plt.plot(epochs, val_list_B, label='Sol B')
plt.plot(epochs, val_list_C, label='Sol C')
plt.plot(epochs, val_list_D_H2, label='Sol D, H.2')
plt.plot(epochs, val_list_D_C2, label='Sol D, C.2')
plt.plot(epochs, val_list_D_F2, label='Sol D, F.2')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Acc')
plt.show()

# graph by time
plt.plot(time_list_B, val_list_B, label='Sol B')
plt.plot(time_list_C, val_list_C, label='Sol C')
plt.plot(time_list_D_H2, val_list_D_H2, label='Sol D, H.2')
plt.plot(time_list_D_C2, val_list_D_C2, label='Sol D, C.2')
plt.plot(time_list_D_F2, val_list_D_F2, label='Sol D, F.2')
plt.legend()
plt.xlabel('Time (sec)')
plt.ylabel('Acc')
plt.show()
```
